```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Fetch AMD data from yfinance
ticker = "btc-usd"
data = yf.download(ticker, start="2023-01-01", end="2025-01-01")

# Add a column for percentage change in volume compared to the previous day
data['Volume_Change'] = data['Volume'].pct_change()

# Filter for days where the volume increased significantly (e.g., by more than 100%)
x = 0.9  # 100% increase
significant_days = data[data['Volume_Change'] > x].index

# Plot the stock price with significant volume spikes highlighted
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label="Close Price", color='blue', linewidth=1.5)

# Scatter plot for significant days (highlighting them in red)
plt.scatter(significant_days, data.loc[significant_days, 'Close'], color='red', label="Significant Volume Spike", s=50)

plt.title(f"{ticker} Stock Price (Last 1 Year) with Volume Spikes")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.legend()
plt.grid(True)
plt.show()
```
